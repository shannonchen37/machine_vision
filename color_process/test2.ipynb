```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 读取原始图像
img = cv2.imread('./interpolation/Snoopy-small.jpg')

# 目标图像大小
dst_h, dst_w = 800, 600

# 原始图像大小
src_h, src_w = img.shape[:2]

# 计算横向和纵向上的缩放比例
ratio_h = dst_h / src_h
ratio_w = dst_w / src_w

# 创建目标图像数组
dst = np.zeros((dst_h, dst_w, 3), dtype=np.uint8)

# 最近邻插值
for i in range(dst_h):
    for j in range(dst_w):
        src_x = min(max(round(j / ratio_w), 0), src_w - 1)
        src_y = min(max(round(i / ratio_h), 0), src_h - 1)
        dst[i, j] = img[int(src_y), int(src_x)]
```

```python
import numpy as np
import cv2
import os
```

```python
# 加载图像并转换为矩阵
images = []
for filename in os.listdir("noise"):
    if filename.endswith(".jpg"):
        img = cv2.imread(os.path.join("noise", filename), 0)
        images.append(img)
images
```

```python
# 对所有图像矩阵进行逐元素求和
sum_matrix = np.zeros_like(images[0])
for img in images:
    sum_matrix += img

# 求取平均值矩阵
mean_matrix = sum_matrix / len(images)

# 对每个原始图像进行去噪增强
for i, img in enumerate(images):
    filtered_img = img - mean_matrix

    # 保存去噪增强后的图像
    cv2.imwrite(os.path.join("./", "filtered_img_{}.jpg".format(i)), filtered_img)
```

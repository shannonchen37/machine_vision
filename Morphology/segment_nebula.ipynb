```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 读取原始图像
img = cv2.imread('./star.png', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
# 定义结构元素
kernel1 = np.ones((3,3), np.uint8)
kernel2 = np.ones((5,5), np.uint8)
kernel3 = np.ones((7,7), np.uint8)

# 进行开操作
opening1 = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel1)
opening2 = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel2)
opening3 = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel3)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].imshow(img, cmap='gray')
axs[0, 0].set_title('origin image')
axs[0, 0].axis('off')
axs[0, 1].imshow(opening1, cmap='gray')
axs[0, 1].set_title('Kernel = 3')
axs[0, 1].axis('off')
axs[1, 0].imshow(opening2, cmap='gray')
axs[1, 0].set_title('Kernel = 5')
axs[1, 0].axis('off')
axs[1, 1].imshow(opening3, cmap='gray')
axs[1, 1].set_title('Kernel = 7')
axs[1, 1].axis('off')
# 调整子图间距
plt.tight_layout()
plt.show()
```

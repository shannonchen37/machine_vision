```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 加载图像
img = cv2.imread('x-rayblog-gausian_blur_d4.jpg')
```

```python
# 定义拉普拉斯算子
kernel1 = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
kernel2 = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
```

```python
# 定义卷积函数
def convolution(image, kernel):
    output = np.zeros_like(image, dtype=float)
    for i in range(1, image.shape[0]-1):
        for j in range(1, image.shape[1]-1):
            output[i, j] = np.sum(image[i-1:i+2, j-1:j+2] * kernel)
    output = (output - np.min(output)) / (np.max(output) - np.min(output)) * 255
    output = output.astype(np.uint8)
    return output
```

```python
output1 = convolution(img,kernel1)
output2 = convolution(img,kernel2)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 20))
axs[0].imshow(img, cmap='gray')
axs[0].set_title('origin image')
axs[1].imshow(output1, cmap='gray')
axs[1].set_title('Sharpened Image with kernel1')
axs[2].imshow(output2, cmap='gray')
axs[2].set_title('Sharpened Image with kernel2')
plt.show()
```

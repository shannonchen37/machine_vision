```python
import cv2
import numpy as np

# Read the image and convert it to grayscale
img = cv2.imread('body_x_ray.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Laplacian filter to the grayscale image
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
```

```python
# Apply Sobel filter to the grayscale image
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1)

# Combine the results of the horizontal and vertical Sobel operators
sobel = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)
```

```python
# Combine Laplacian and Sobel images
sharp = cv2.addWeighted(laplacian, 0.5, sobel, 0.5, 0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(sharp,cmap='gray')
plt.title('sharpen')
plt.show()
```

```python
final = gray - sharp
```

```python
plt.imshow(final,cmap='gray')
plt.title('final')
plt.show()
```

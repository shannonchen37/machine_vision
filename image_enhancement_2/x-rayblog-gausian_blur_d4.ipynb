```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 读取图像
img = cv2.imread('x-rayblog-gausian_blur_d4.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# 创建拉普拉斯核
laplacian_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
```

```python
# 使用OpenCV函数cv2.Laplacian()进行卷积
laplacian_img = cv2.Laplacian(img, cv2.CV_8U, ksize=3)
```

```python
# 显示原始图像和锐化图像
plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(laplacian_img, cmap='gray'), plt.title('Sharpened')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
sharpened_img = img - laplacian_img
```

```python
# 显示原始图像和锐化图像
plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sharpened_img, cmap='gray'), plt.title('Sharpened')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2

# 读取灰度图像
img = cv2.imread('./2.jpg', 0)

# 应用水平方向的Sobel滤波器
sobelx = cv2.Sobel(img, cv2.CV_16S, 1, 0, ksize=3)

# 对Sobel滤波器输出进行阈值处理
thresh = cv2.threshold(sobelx, 50, 255, cv2.THRESH_BINARY)[1]

# 对二值化图像进行形态学操作，去除噪声和不需要的轮廓
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (21, 7))
morph = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
```

```python
import matplotlib.pyplot as plt

# 显示结果
plt.subplot(221), plt.imshow(img, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(sobelx, cmap='gray')
plt.title('Sobel'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(thresh, cmap='gray')
plt.title('Threshold'), plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(morph, cmap='gray')
plt.title('Morphology'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import numpy as np
# 定义水平方向的卷积核
kernel = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]], dtype=np.float32)

# 对图像进行卷积操作，得到水平方向上的梯度图像
grad = cv2.filter2D(img, -1, kernel)

# 对梯度图像进行阈值处理，得到二值化图像
thresh = cv2.threshold(grad, 50, 255, cv2.THRESH_BINARY)[1]

# 对二值化图像进行形态学操作，去除噪声和不需要的轮廓
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (21, 7))
morph = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

# 显示结果
plt.subplot(221), plt.imshow(img, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(grad, cmap='gray')
plt.title('Gradient'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(thresh, cmap='gray')
plt.title('Threshold'), plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(morph, cmap='gray')
plt.title('Morphology'), plt.xticks([]), plt.yticks([])
plt.show()
```
